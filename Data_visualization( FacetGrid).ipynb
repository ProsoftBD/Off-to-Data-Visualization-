```python
#Import necessary module
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Initialize FacetGrid 
sns.FacetGrid(tips, row='smoker', col='time')
```

```python
#Check the columns 
tips.head()
```

```python
tips.tail()
```

```python
#Now apply map function and lets see what happen 

x= sns.FacetGrid(tips, row='smoker', col='time')
x= x.map(plt.hist,'total_bill')
```

```python
#Change the color of the Facetgrid 
x= sns.FacetGrid(tips, row='smoker', col='time')
x= x.map(plt.hist,'total_bill', color='green')
```

```python
#Change the number of bin 
x= sns.FacetGrid(tips, row='smoker', col='time')
x= x.map(plt.hist,'total_bill', color='blue', bins=15)
```

```python
#Using matplotlib scatter function
x= sns.FacetGrid(tips, row='smoker', col='time')
x= x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x= sns.FacetGrid(tips, row='smoker', col='time', hue='smoker')
x= x.map(sns.regplot, 'total_bill', 'tip')
```

```python
#Above plot, hue is not showing,. In that case, we need to add legend function to perform the action.

x= sns.FacetGrid(tips, row='smoker', col='time', hue='smoker')
x= (x.map(sns.regplot, 'total_bill', 'tip').add_legend()) #Just comapre the above plot and below. Here smoker hue is showing after add legend function.
```

```python
#Using boxplot and see what happen
x= sns.FacetGrid(tips, col='day')
x= x.map(sns.boxplot, 'total_bill','time')
```

```python
#Increase size 
x= sns.FacetGrid(tips, col='day', size=6, aspect=.7, col_order=['Fri','Sat','Sun','Thur'])
x= x.map(sns.boxplot, 'total_bill','time', color='red')
```
