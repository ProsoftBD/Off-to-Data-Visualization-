```python
#load necessary module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#load dataset
```

```python
tips= sns.load_dataset('tips')
```

```python
sns.violinplot(x= tips['tip'])
#tips.head()
```

```python
sns.violinplot(x=tips['total_bill'])
```

```python
#make violinplot for two varibale 
sns.violinplot(x='size', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='smoker', y='total_bill', data=tips)
```

```python
sns.violinplot(x='smoker', y='total_bill', data=tips,hue='sex')
```

```python
sns.violinplot(x='smoker', y='total_bill', data=tips,hue='sex', palette='husl')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='smoker', split= True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='smoker', split= True, palette='coolwarm')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='smoker', split= True, palette='coolwarm', order=['Sat', 'Sun', 'Thur', 'Fri'])
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='smoker', split= True, palette='coolwarm', order=['Sat', 'Sun', 'Thur', 'Fri'])
```

```python
#Make together that seems like cool and easy to find out the difference.

sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', inner='stick', scale='count', split= True)
```

```python
#Make together that seems like cool and easy to find out the difference.

sns.violinplot(x='day', y='total_bill', data=tips,scale_hue=False, hue='sex', inner='stick', scale='count', split= True)
```
