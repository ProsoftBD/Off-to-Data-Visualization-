```python
#Import necessary module
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
import numpy as np
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
```

```python
x_axis, y_axis= np.random.multivariate_normal(mean,cov, size=40).T
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
#Use shade and color for X axis 
sns.kdeplot(x_axis, shade= True, color='green')
```

```python
#Use shade and color for Y axis
sns.kdeplot(y_axis, shade= True, color='red')
```

```python
sns.kdeplot(x_axis, y_axis) #For two axis
```

```python
sns.kdeplot(x_axis, y_axis, shade= True)
```

```python
#Using color map
sns.kdeplot(x_axis, y_axis, cmap='RdBu')
```

```python
#Using color map
sns.kdeplot(x_axis, y_axis, cmap='coolwarm')
```

```python
sns.kdeplot(x_axis, y_axis, n_levels=35, cmap='coolwarm')
```

```python
#Try by vertical and horizontal axis\
sns.kdeplot(x_axis, vertical= True)
```

```python
#Import Iris dataset to test the kde plot
```

```python
iris= sns.load_dataset('iris')

```

```python
setosa= iris.loc[iris.species=='setosa']
versicolor= iris.loc[iris.species=='versicolor']

```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width)
sns.kdeplot(versicolor.petal_length, versicolor.petal_width)
```

```python
#Using color map
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='coolwarm')
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap='RdBu', shade= True)
```

```python
#Change the position and lets whether petal width is X AXIS
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='coolwarm')
sns.kdeplot(versicolor.petal_width,versicolor.petal_length, cmap='RdBu')
```
