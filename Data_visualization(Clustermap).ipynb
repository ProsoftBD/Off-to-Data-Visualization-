```python
#load necessary module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#Load dataset 
flights= sns.load_dataset('flights')
```

```python
#Check how many attribute are there in this dataset 

flights.head()
```

```python
#Now we will pivot our flights data
flights= flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
#Use color map and lets see what happen.

sns.clustermap(flights, cmap='coolwarm')
```

```python
#use linewidth 
sns.clustermap(flights, cmap='coolwarm', linewidth=1)
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidth=1.5)
```

```python
#Lets see by define figure size
sns.clustermap(flights, cmap='Blues_r', linewidth=1.5, figsize=(8,6))
```

```python
#Standardrize across col or row
sns.clustermap(flights, standard_scale=1)
```

```python
#Normalizing our dataset 
sns.clustermap(flights, z_score=0)
```

```python
sns.clustermap(flights, z_score=1)
```
