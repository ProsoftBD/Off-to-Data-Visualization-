```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline 
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
sns.stripplot(x=tips['tip'])
```

```python
sns.stripplot(x=tips['tip'], color='green')
```

```python
sns.stripplot(x=tips['total_bill'])
```

```python
tips.head()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='total_bill',y='day', data=tips, jitter= 1.3, linewidth=1.2)
```

```python
sns.stripplot(x='total_bill',y='day', data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips,hue='smoker', jitter= True,\
              split= True,palette='winter_r', order=['Fri', 'Sta', 'Sun', 'Thur'])
```

```python
sns.stripplot(x='sex', y='tip', data=tips, order=['Female', 'Male'])
```

```python
sns.stripplot(x='sex', y='tip', data=tips, marker='D', order=['Female', 'Male'])
```

```python
sns.stripplot(x='sex', y='tip', data=tips, marker='D',size=15, order=['Female', 'Male'])
```

```python
sns.stripplot(x='sex', y='tip', data=tips, marker='D',size=15, hue='sex', \
             edgecolor='gray', alpha=0.3, jitter= True)
```

```python
sns.stripplot(x='sex', y='tip', data=tips, marker='D',size=15, hue='sex', \
             edgecolor='gray', alpha=0.3, jitter= True, palette='husl')
```

```python
sns.stripplot(x='sex', y='tip', data=tips)
sns.boxplot(x='sex', y='tip', data=tips)
```

```python
sns.stripplot(x='sex', y='tip', data=tips, color='1')
sns.boxplot(x='sex', y='tip', data=tips)
```

```python
sns.stripplot(x='sex', y='tip', data=tips, jitter= True)
sns.violinplot(x='tip', y='day', data=tips, color='0.6')
```
