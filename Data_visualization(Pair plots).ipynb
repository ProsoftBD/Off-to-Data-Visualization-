```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
%matplotlib inline 
```

```python
#Now we are going to import iris dataset 
```

```python
iris= sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
#Now import tips dataset 
tips= sns.load_dataset('tips')
sns.pairplot(tips)
```

```python
#Now pass the hue attribute 
sns.pairplot(iris, hue='species')
```

```python
#Now pass the palette attribute
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, hue='species', palette='Blues_d')
```

```python
sns.pairplot(iris, hue='species', markers=['o','D','s'])
```

```python
sns.pairplot(iris, vars=['sepal_length','sepal_width'])
```

```python
#Define the size
sns.pairplot(iris, size=6,vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris, x_vars=['petal_length','petal_width'],\
            
             y_vars=['sepal_length','sepal_width'],\
             palette='husl',hue='species'
            )
```

```python
sns.pairplot(iris, diag_kind='kde', palette='husl', hue='species')
```

```python
sns.pairplot(iris, kind='reg')
```

```python
sns.pairplot(iris, kind='reg', palette='husl', hue='species')
```
