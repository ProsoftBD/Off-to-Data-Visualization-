```python
#Import necessary module
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
import numpy as np
```

```python
#Regression plot is a kind of plot which allows us to see the relationahsip in the variable 
```

```python
#load our tips dataset
tips= sns.load_dataset('tips')
tips.head()
```

```python
#Plot into the regplot

sns.regplot(x='total_bill',y='tip', data=tips)
```

```python
#use color 
sns.regplot(x='total_bill',y='tip', data=tips, color='green')
```

```python
#Now use multivariate normal method to create data to crearing a normally distrubuted data
```

```python
mean=[2,5]
cov=[[1.1,0.4],[2.2,3]]
```

```python
x_value, y_value= np.random.multivariate_normal(mean, cov, 100).T #Lets see the nice regression line.
sns.regplot(x=x_value, y=y_value)
```

```python
sns.regplot(x=x_value, y=y_value, color='purple')
```

```python
series_x_value= pd.Series(x_value, name='Var-X')
series_y_value= pd.Series(y_value, name='Var-Y')

sns.regplot(x= series_x_value, y= series_y_value)
```

```python
sns.regplot(x= series_x_value, y= series_y_value, color='gray', marker='d')
```

```python

sns.regplot(x='total_bill', y='tip', data= tips,marker='D',\
           scatter_kws={'color':'blue'},\
            line_kws={'color':'red'}
           )
```

```python
#change market to + singed and change the color of line width 
sns.regplot(x='total_bill', y='tip', data= tips,marker='+',\
           scatter_kws={'color':'gray'},\
            line_kws={'color':'black','linewidth':2.5}
           )
```

```python
#for another value
#change market to + singed and change the color of line width 
sns.regplot(x=x_value, y=y_value, data= tips,marker='D',\
           scatter_kws={'color':'gray'},\
            line_kws={'color':'black','linewidth':2.5}
           )
```

```python

sns.regplot(x= 'size', y= 'total_bill',data=tips, x_jitter=0.5)
```

```python
#See tha mean of value 

sns.regplot(x= 'size', y= 'total_bill',data=tips, x_estimator= np.mean)
```
