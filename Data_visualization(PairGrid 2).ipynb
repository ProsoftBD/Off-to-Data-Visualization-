```python
#import module 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline 
import seaborn as sns
```

```python
iris= sns.load_dataset('iris')
```

```python
x= sns.PairGrid(iris, vars=['petal_length', 'petal_width'])
x= x.map(plt.scatter)
```

```python
#Change diagoal funcion
x= sns.PairGrid(iris,hue='species', vars=['petal_length', 'petal_width'])
x= x.map_diag(plt.hist, edgecolor='black')#edge color has changed 
x= x.map_offdiag(plt.scatter, edgecolor='red') #edge color has changed for the scatter plot 
x= x.add_legend()
```

```python
x= sns.PairGrid(iris, x_vars=['petal_length', 'petal_width'],\
                y_vars=['sepal_length', 'sepal_width'])
x= x.map(plt.scatter)
```

```python
x= sns.PairGrid(iris)
x= x.map_diag(plt.hist)
x= x.map_upper(plt.scatter)
x= x.map_lower(sns.kdeplot)
```

```python
x= sns.PairGrid(iris, hue='species', palette='coolwarm')
x= x.map_diag(plt.hist)
x= x.map_upper(plt.scatter)
x= x.map_lower(sns.kdeplot)
x= x.add_legend()
```

```python
x= sns.PairGrid(iris, hue='species',palette='RdBu', hue_kws={'marker':['D','s','+']})
x= x.map_diag(plt.hist)
x= x.map_upper(plt.scatter, s=30)
x= x.map_lower(sns.kdeplot)
x= x.add_legend()
```
