```python
#Import necessary module
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
import numpy as np
```

```python
#Load dataset
tips= sns.load_dataset('tips')
tips.head()
```

```python
#load the iris dataset

iris= sns.load_dataset('iris')
iris.head()
```

```python
#Using joint plot 

sns.jointplot(x='total_bill', y='tip', data= tips)
```

```python
#Check by iris dataset 
sns.jointplot(x='sepal_length', y='sepal_width', data= iris)
```

```python
#Now make regression line for scatter plot to see the regression line 
sns.jointplot(x='total_bill', y='tip', kind= 'reg', data= tips)
```

```python
#Change the color of the joint polot
sns.jointplot(x='total_bill', y='tip', kind= 'reg', data= tips, color='green')
```

```python
#Hexagonal presentation

sns.jointplot(x='total_bill', y='tip', data= tips, kind='hex')
```

```python
#Lets see total bill vs tip 
sns.jointplot(x='total_bill', y='tip', data= tips, kind='kde')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data= iris, kind='kde')
```

```python
#Use color attribute
sns.jointplot(x='sepal_length', y='sepal_width', data= iris, kind='kde', color='red')
```

```python
from scipy.stats import spearmanr 
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind='kde', stat_func= spearmanr )
```

```python
sns.jointplot(x='size', y='total_bill', data= tips, stat_func= spearmanr, color='blue' )
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, ratio=1, size=6 )
```
